#### I am going to use k-means clustering. Will divide the schools into 4 categories

```python
# Noralize the dataset
from sklearn.preprocessing import StandardScaler

X = hs_df4.values
cluster_dataset = StandardScaler().fit_transform(X)
```

```python
from sklearn.cluster import KMeans 

num_clusters = 4

k_means = KMeans(init="k-means++", n_clusters=num_clusters, n_init=12)
k_means.fit(cluster_dataset)
labels = k_means.labels_

print(labels)
```

```python
import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

results = requests.get(url).json()
results
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories'] 
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name'] 
```

```python
university = results['response']['groups'][0]['items']
    
nearby_university = json_normalize(university) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_university =nearby_university.loc[:, filtered_columns]

# filter the category for each row
nearby_university['venue.categories'] = nearby_university.apply(get_category_type, axis=1)
```

```python
# clean columns
nearby_university.columns = [col.split(".")[-1] for col in nearby_university.columns]

nearby_university.head()
```
